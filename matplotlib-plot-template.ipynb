```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0,2*np.pi,100)
ys = np.sin(xs)
zs = np.cos(xs)
```

```python
large = 40; med = 50; small = 20
params = {'axes.titlesize': med,
          'axes.titlepad' : med,
          'legend.fontsize': med,
          'axes.labelsize': med ,
          'axes.titlesize': med ,
          'xtick.labelsize': med ,
          'ytick.labelsize': med ,
          'figure.titlesize': med}
plt.rcParams['text.usetex'] = True
plt.rcParams.update(params)
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(15)
plt.title(r"A plot of sine function")
plt.plot(xs,ys, linewidth=5,color="tab:red", label = r"$f(x) = \sin(x)$")
plt.scatter(xs,zs,s=100,color="tab:green", edgecolor ="k", marker = "s",label=r"$g(x) = \cos(x)$")
plt.grid(linestyle='--', alpha=0.5)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$",labelpad = 10)
plt.legend()
fig.tight_layout(pad=1.0)
ax = fig.gca()
for spine in ax.spines.values():
    spine.set_linewidth(3)
plt.gcf().subplots_adjust(bottom=0.2)

plt.tick_params(axis="x", direction="inout", length=20, width=3, color="k")
plt.tick_params(axis="y", direction="inout", length=20, width=3, color="k")

plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],
           [r"$0$",r"$\pi/2$",r"$\pi$",r"$3\pi/2$",r"$2\pi$"])
plt.yticks([-1,0,1],[r"$-1$",r"$0$",r"$1$"])
plt.show()
```
